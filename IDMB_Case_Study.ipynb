## change gender male as 1 and female as 0?

```python
def encode(gender):
   if gender == 'Male':
      return 1
   else:
      return 0
```

## Which director is most productive?

```python
data = df.groupby('director_name').agg({'title':'count','year':['max','min']})
```

```python
data
```

```python
data['year_active'] = data['year','max']-data['year','min']
```

```python
data
```

```python
data['productivity'] = data['title','count']/data['year_active']
```

```python
data.head()
```

```python
final_data = data.reset_index(0)
```

```python
final_data.sort_values(by ='productivity',ascending = False)[:1][['director_name','productivity']]
```
